# **Pearson (item-item)**

```python
tr = ratings_with_name.pivot_table(index='user_id',columns='title',values='rating')
```

```python
tr
```

```python
def avg_calcu2(col):
  df = pd.DataFrame(tr, columns=[col])
  filtered_df = df.dropna()

  # Calculate the average
  average = np.mean(filtered_df[col])
  return average
```

```python
def avg_calcu1(title):
  df = pd.DataFrame(tr, columns=[title])
  filtered_df = df.dropna()

  # Calculate the average
  average = np.mean(filtered_df[title])
  return average
```

```python
def pearson(title):
    given_title = tr[title]
    corr_summary = []

    for col in tr.columns:
        if col == title:
            continue
        avg_1 = avg_calcu1(title)
        avg_2 = avg_calcu2(col)

        compare_column = tr[col]
        movie_common_1 = []
        movie_common_2 = []

        for i in range (len(given_title)):
          if pd.notnull(given_title.iloc[i]) and pd.notnull(compare_column.iloc[i]):
            movie_common_1.append((given_title.iloc[i]))
            movie_common_2.append((compare_column.iloc[i]))

        # common_user give me index value

      
        if len(movie_common_1) >= 2:  # Consider movies with at least 2 common users
            
            num = ((movie_common_1 - avg_1) * (movie_common_2 - avg_2)).sum()
            den = np.sqrt(((movie_common_1 - avg_1) ** 2).sum()) * np.sqrt(((movie_common_2 - avg_2) ** 2).sum())
            
            if den != 0:  # Avoid division by zero
                pearson_coeff = num / den
                corr_summary.append((col, pearson_coeff))

    corr_summary.sort(key=lambda x: x[1], reverse=True)
    
    filtered_corr_summary = []
    count = 0

    for col, coeff in corr_summary:
     if len(filtered_corr_summary) == 10:
        break
     filtered_corr_summary.append((col, coeff))
    return filtered_corr_summary
```

# **USER-USER**

```python
us = ratings_with_name.pivot_table(index='title',columns='user_id',values='rating')
```

```python
us
```

```python
def avg_calcu1(id):
  df = pd.DataFrame(us, columns=[id])
  filtered_df = df.dropna()

  # Calculate the average
  average = np.mean(filtered_df[id])
  return average
```

```python
def avg_calcu2(col):
  df = pd.DataFrame(us, columns=[col])
  filtered_df = df.dropna()

  # Calculate the average
  average = np.mean(filtered_df[col])
  return average
```

```python
def user_pearson(id):
    given_user = us[id]
    corr_summary = []

    for col in us.columns:
        if col == id:
            continue
        avg_1 = avg_calcu1(id)
        avg_2 = avg_calcu2(col)

        compare_ratings = us[col]
        user_common_1 = []
        user_compare_common = []

        compare_ratings = us[col]
        for i in range(len(given_user)):
            if pd.notnull(given_user.iloc[i]) and pd.notnull(compare_ratings.iloc[i]):
                user_common_1.append((given_user.iloc[i]))
                user_compare_common.append((compare_ratings.iloc[i]))

        # common_user give me index value
        if len(user_common_1) >= 2:  # Consider movies with at least 2 common users
            num = ((user_common_1 - avg_1) * (user_compare_common - avg_2)).sum()
            den = np.sqrt(((user_common_1 - avg_1) ** 2).sum()) * np.sqrt(((user_compare_common - avg_2) ** 2).sum())
            
            if den != 0:  # Avoid division by zero
                pearson_coeff = num / den
                corr_summary.append((col, pearson_coeff))

    corr_summary.sort(key=lambda x: x[1], reverse=True)

    filtered_corr_summary = []
    count = 0

    for col, coeff in corr_summary:
        if len(filtered_corr_summary) == 1:
            break
        filtered_corr_summary.append((col, coeff))
    
    # Retrieve top 10 movies rated greater than or equal to 3 by the resultant user
    rated_top_movies = []
    for col,coeff in filtered_corr_summary:
        user_ratings = us[col]
        user_top_movies = user_ratings[user_ratings >= 3].index[:10]
        rated_top_movies.extend(user_top_movies)
    
    return rated_top_movies
```
